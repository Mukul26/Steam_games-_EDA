### Since the stakeholders have asked to find out reviewers with the highest votes_helpful so they can be incentivize with early releases of products we will use the groupby function

```python
top_Helpful_votes = df1.groupby('author.steamid')['votes_helpful'].sum().reset_index().sort_values('votes_helpful',ascending=False)
top_Helpful_votes
```

```python
#changing column name of from votes_helpful to total_helpful_votes
top_Helpful_votes.rename(columns = {'votes_helpful':'total_helpful_votes'},inplace=True)
```

### Since stakeholders have asked us to be careful about one hit wonders, we will look into the count of reviews of top people with most helpful_votes 

```python
review_count = df1.groupby('author.steamid')['review_id'].count().reset_index().sort_values('review_id',ascending=False)
review_count 
```

```python
# again we will change the columns name from review_id to review_id_count
review_count.rename(columns={'review_id':'review_id_count'},inplace=True)
```

```python
review_count.head()
```

## Now we will merge review_count dataframe with  top_helpful_votes

```python
df2 = top_Helpful_votes.merge(review_count,on='author.steamid',how='left')
df2
```

## NOTE : We should keep in mind that review_id_count only represents the number of record we have in ths dataset about the particular author.steamid, you should not assume they only commented  that many times. author.num_reviews is the true representation about the number of reviews someone from a particular steam.authorid have on the platform

```python
df3 = review_count.merge(top_Helpful_votes,on='author.steamid',how='left')
df3
```

```python
sb.set(rc={"figure.figsize":(13, 8)}) #width=3, #height=4
sb.histplot(data=df3,x='review_id_count',binwidth=1)
```

Above histogrm shows most people only have one review so the stakeholders expection about incentivizing active users with count of reviews is not possible since more than 85% of the people have only one review.
